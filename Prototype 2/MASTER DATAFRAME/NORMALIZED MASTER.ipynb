```python
import pandas as pd
```

```python
df = pd.read_csv("MASTER_system_and_weather.csv")
```

```python
df.tail(30)
```

```python
df.columns
```

```python
columns_to_normalize = ['Generated (kWh)','System Size (kW)', 'Solar_Irradiance', 'Temperature_2m',
       'Cloud_Cover', 'Efficiency']
```

```python
for col in columns_to_normalize: 
    max_value = df[col].max()
    print(f"{df[col].name}: {max_value}")
    if (max_value != 0):
        df[col] = df[col] / max_value
    else:
        df[col] = 0
```

```python
df.tail(30)
```

```python
df.to_csv("NORMALIZED_MASTER.csv", index=False)
```
